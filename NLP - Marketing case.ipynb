There are more neutral words than positive.

The compound score is the sum of positive, negative & neutral scores which is then normalized between -1(most extreme negative) and +1 (most extreme positive).

The more Compound score closer to +1, the higher the positivity of the text.

```python
def review_rating(string):
    sid_1= sid.polarity_scores(review)
    if sid_1["compound"] ==0:
        return "Neutral"
    elif sid_1["compound"]>0:
        return "Positive"
    else:
        return "Negative"
```

# Topic Modeling

```python
## Build pipelines to vectorize the data, then train and fit a model
# Now that we have sets to train and test, we'll develop a selection of pipelines, each with a different model
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
```

```python
dtm = tfidf.fit_transform(x)
```

### Non-negative Matrix Factorization

#### TASK: Using Scikit-Learn create an instance of NMF

```python
from sklearn.decomposition import NMF
nmf_model = NMF(n_components=5,random_state=42)
```

```python
nmf_model.fit(dtm)
```

#### TASK: Print our the top 15 most common words 

```python
single_topic = nmf_model.components_
```

```python
for index,topic in enumerate(nmf_model.components_):
    print(f'THE TOP 15 WORDS FOR TOPIC #{index}')
    print([tfidf.get_feature_names()[i] for i in topic.argsort()[-15:]])
    print('\n')
```
